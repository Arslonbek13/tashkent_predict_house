```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

X_train = train_set.drop("price", axis=1)
y = train_set["price"].copy()
letter_name = ['district', 'location']
X_num = X_train.drop(letter_name, axis=1)
```

```python
column_indexes = df.columns
column_indexes
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, per_room_size=True):
        self.per_room_size = per_room_size

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        rooms_ix, size_ix, level_ix, max_levels_ix = 0, 1, 2, 3  # Adjust these indices based on your DataFrame

        per_size = X[:, size_ix] / X[:, rooms_ix]
        per_level = X[:, max_levels_ix] / X[:, level_ix]

        if self.per_room_size:
            return np.c_[X, per_size, per_level]
        else:
            return np.c_[X, per_level]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline([
          ('imputer', SimpleImputer(strategy='median')),
          ('attribs_adder', CombinedAttributesAdder(per_room_size = True)),
          ('std_scaler', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(X_num)
cat_attribs = ['district']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])
```

```python
X_prepared = full_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression


LR_model = LinearRegression()


LR_model.fit(X_prepared, y)
```
